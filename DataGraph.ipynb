```python
import matplotlib.pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67. , 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.plot(years, gdp, color='red', marker='p', linestyle='--')
plt.grid(True, color='gray', alpha=0.5, linestyle='--')

plt.title("GDP per capita")

plt.ylabel=('$')
plt.savefig("gdp_per_capita.png", dpi=600)
plt.show()
```

```python
x = [x for x in range(-20, 20)]
```

```python
y1 = [2*t for t in x]
y2 = [t**2 + 5 for t in x]
y3 = [-t**2 -5 for t in x]
```

```python
plt.plot(x, y1, 'r-', x, y2, 'b--', x, y3, 'g--', 0, 0, 'o')
plt.legend()
plt.grid(True, color='gray', alpha=0.5, linestyle='--')
plt.axis([-30, 30, -30, 30])
plt.savefig("function_graph.png", dpi=600)
plt.show()
```

```python
x = [x for x in range(20)]
y = [x**2 for x in range(20)]
z = [x**3 for x in range(20)]

plt.plot(x, x, label='linear')
plt.plot(x, y, label='quadratic')
plt.plot(x, z, label='qubic')

plt.title("My plot")
plt.legend()
plt.show()
```
